```python
import numpy as np
```

```python
I = range(-5, 5+2)
```

```python
END = 6

def p(a,i,j,rho=0.5):
    if a == 0:
        if i == -5:
            return float( (1-rho) * (j==-5)  + rho * (j==-4) )
        if i == 5:
            return float( rho * (j==5) + (1-rho) * (j==4) )
        if i == END:
            return (j==6)
        else:
            return float( (1-rho) * (j==(i-1)) + rho * (j==(i+1)) )
    if a == 1:
        return (j==6)
```

```python
def P(a):
    pa = np.empty((12,12))
    for i in range(-5,5+2):
        for j in range(-5,5+2):
            pa[5+i,5+j] = p(a,i,j)
    return pa
```

```python
P0 = P(0)
P1 = P(1)
P0,P1
```

```python
def r(a,i):
    if a == 0 and i != 6:
        return i
    else:
        return 0
```

```python
alpha = 0.95

def T(alpha, f, v, i):
    return r(f[5+i],i) + alpha * sum([
        p(f[5+i], i, j) * v[5+j]
        for j in I
    ])
```

```python
v0 = [0 for _ in I]
f0 = 5*[1] + 7*[0]
```

```python
v_prev = np.array([T(alpha, f0, v0, i) for i in I])
v_next = np.empty(len(I))

# wider printing
np.set_printoptions(linewidth=400)

epsilon = 0.0001
diff = 1
while diff > epsilon:
    v_next = np.array([T(alpha, f0, v_prev, i) for i in I])
    M = max(v_next-v_prev)
    m = min(v_next-v_prev)
    diff = abs(m-M)
    v_prev = v_next
    print(v_prev)
```
