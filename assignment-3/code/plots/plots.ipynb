```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
suff = '_chart_data.tsv'
files = ['reward', 'reward_sma', 'reward_eval', 'loss', 'epsilon', 'avg_max_q']
dfs = {file: pd.read_csv(file + suff, sep='\t', header=0).fillna(0).drop('run_id', axis=1) for file in files}
```

```python
# clip outliers
dfs['loss']['loss'] = dfs['loss']['loss'].clip(upper=1)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(20, 10))
for i, (name, df) in enumerate(dfs.items()):
    ax = axs[i // 3, i % 3]
    df.plot(x='step', y=1, kind='line', ax=ax)
    plt.xlabel('step')
    # plt.show()

plt.savefig('evaluation.png')
```
